```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib 

#matplotlib.rcParams.update({'font.size': 14})
```

```python

tasks = ['Task1', 'Task2', 'Task3', 'Task4', 'Task5', 'Task6', 'Task7', 'Task8', 'Task9', 'Task10']
start_months = [0, 6, 12, 16, 18,24,28, 30, 38, 40]
duration = [6, 6,6, 2,6,6,2,8, 2, 6]
end_months = [sum(x) for x in zip(start_months,duration)]
completion = [1, 1,.5,0,0,0,0,0,0, 0]
```

```python
df = pd.DataFrame(list(zip(tasks, start_months, duration, end_months, completion))
                , columns =['tasks', 'start_months', 'duration', 'end_months', 'completion'])
df['current_num'] = (df.duration * df.completion)
```

```python
fig, ax = plt.subplots(1, figsize=(16,6))
ax.barh(df.tasks, df.current_num, left=df.start_months, alpha=0.5, color = 'r')
ax.barh(df.tasks, df.duration, left=df.start_months, alpha=0.5, color = 'r')

for idx, row in df.iterrows():
    ax.text(row.end_months+0.1, idx, f"{int(row.completion*100)}%", va='center', alpha=0.5)
xticks = [0,6,12,18,24,30,36,42,48]#start_months
xticks_labels = [0,6,12,18,24,30,36,42,48]#start_months
xticks_minor = np.arange(0, 46, 1)
ax.set_xticks(xticks)
ax.set_xticks(xticks_minor, minor=True)
ax.set_xticklabels(xticks_labels)
ax.set_xlabel('Months')
plt.savefig('ganttchart.png', dpi=300, bbox_inches='tight')
plt.show()
```
